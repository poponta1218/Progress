```julia
using Plots
L=10*10^-10
A=sqrt(2/L)
x=range(0,L,length=10000)
f1=plot(x,x->A*sin(/L*x),label=false,title="n=1")
f2=plot(x,x->A*sin(2*/L*x),label=false,title="n=2")
plot(f1,f2,layout=(2,1),link=:x,ylims=(-1.01*A,1.01*A),linewidth=1.5)
```
